```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
def get_sample(x, y, p=0.8):
    assert len(x) == len(y)
    
    y_noise = y * (0.5 + np.random.random(len(x))*1 )
    
    idxs = sorted( random.sample(range(len(x)), k=int(len(x)*p)) )

    return x[idxs], y_noise[idxs]

# Mean Squared Error
def mse(y1, y2):
    return ((y1 - y2)**2).mean()
```

# Funkce sinus

```python
x = np.linspace(0, np.pi*4, 100)
y = np.sin(x)
```

```python
plt.plot(x, y)
```

```python
x_sample, y_sample = get_sample(x, y, p=0.5)
```

```python
plt.plot(x, y, "b")
plt.plot(x_sample, y_sample, "g")
```

### Lineární interpolace

```python
f = interpolate.interp1d(x_sample, y_sample, fill_value="extrapolate")
y_interp = f(x)

plt.plot(x, y, "b")
plt.plot(x, y_interp, "r")
```

```python
mse(y, y_interp)
```

### Kubická spline interpolace

```python
f = interpolate.CubicSpline(x_sample, y_sample)
y_interp = f(x)

plt.plot(x, y, "b")
plt.plot(x, y_interp, "r")
```

```python
mse(y, y_interp)
```

### PCHIP interpolace

```python
f = interpolate.PchipInterpolator(x_sample, y_sample)
y_interp = f(x)

plt.plot(x, y, "b")
plt.plot(x, y_interp, "r")
```

```python
mse(y, y_interp)
```
